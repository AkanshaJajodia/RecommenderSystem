```python

import pandas as pd
```

```python
# reading data to pandas dataframe
df = pd.read_json('reviews_Movies_and_TV_5.json.gz', lines=True)
```

```python
df.head()
```

```python
# extracting columns that is needed
df = df[['reviewerID', 'asin', 'unixReviewTime', 'overall']]

df.columns = ['user', 'item', 'timestamp', 'rating']
```

```python
# renaming the data columns
df.head()
```

```python
# converting the datatype 
df['user_id'] = df['user'].astype("category").cat.codes
df['item_id'] = df['item'].astype("category").cat.codes
```

```python
df.head()
```

```python
# extracting columns that is needed for recommendation
df = df[['user_id', 'item_id', 'rating','timestamp']]
```
